# **6 - Colocar etiquetas para identificar su fuente de origen**

```python
import pandas as pd
from urllib.parse import urlparse

# Supongamos que `df_union` es tu DataFrame combinado con todas las columnas y filas.

# Normalizar las URLs eliminando el prefijo "www."
def normalize_url(url):
    parsed_url = urlparse(url)
    netloc = parsed_url.netloc.replace('www.', '')  # Eliminar "www."
    normalized_url = f"{parsed_url.scheme}://{netloc}{parsed_url.path}"
    return normalized_url

# Aplicar la función de normalización a la columna de URLs en df_union
df_union['normalized_url'] = df_union['url'].apply(normalize_url)

# Crear una tabla de mapeo de dominios a fuentes unificadas
domain_mapping = {
    'theage.com.au': 'The Age',
    'smh.com.au': 'The Sydney Morning Herald',
    'thestar.com.my': 'The Star (Malaysia)',
}

# Función para asignar la fuente basada en el dominio
def map_to_source(url):
    domain = urlparse(url).netloc
    return domain_mapping.get(domain, domain)  # Usar el dominio original si no está en el mapeo

# Aplicar el mapeo a las URLs normalizadas
df_union['source_label'] = df_union['normalized_url'].apply(map_to_source)

# Verificar las primeras filas después de agregar las nuevas columnas
print(df_union[['url', 'normalized_url', 'source_label']].head())

# Opcional: guardar el DataFrame modificado si lo necesitas
df_union.to_csv('estructura_final.csv', index=False)

print("El DataFrame ha sido actualizado con las nuevas columnas.")
```

# 8. Colocar etiquetas para identificar su tema (ayuda de asistente virtual)

```python
import pandas as pd

# Cargar el archivo
file_path = 'estructura_final.csv'  # Reemplaza con la ruta correcta del archivo
df = pd.read_csv(file_path, sep=',')

# Función para clasificar los títulos
def classify_title(title):
    title = str(title).lower()
    if any(keyword in title for keyword in ['ai', 'artificial intelligence', 'chatgpt', 'openai', 'machine learning']):
        return 'AI'
    elif any(keyword in title for keyword in ['tech', 'technology', 'gadget', 'device', 'smartphone', 'google', 'microsoft',
                                              'apple', 'samsung', 'linux', 'hardware', 'software']):
        return 'TECH'
    elif any(keyword in title for keyword in ['gaming', 'game', 'playstation', 'xbox', 'nintendo', 'pc', 'roguelike']):
        return 'GAMING'
    elif any(keyword in title for keyword in ['security', 'cybersecurity', 'hacker', 'data breach', 'encryption', 'privacy']):
        return 'CYBERSECURITY'
    elif any(keyword in title for keyword in ['politics', 'government', 'law', 'election', 'policy']):
        return 'POLITICS'
    elif any(keyword in title for keyword in ['medicine', 'health', 'hospital', 'drug', 'treatment', 'diagnostic', 
                                              'wellness', 'lifestyle', 'movie', 'film', 'tv', 'series', 'netflix', 
                                              'hollywood', 'entertainment']):
        return 'LIFESTYLE'
    elif any(keyword in title for keyword in ['climate', 'environment', 'sustainability', 'energy', 'solar', 'green',
                                              'economy', 'finance', 'stock', 'market', 'investment', 'cryptocurrency',
                                              'culture', 'society', 'social', 'global']):
        return 'SOCIETY'
    elif any(keyword in title for keyword in ['car', 'automobile', 'vehicle', 'electric', 'transport', 'mobility']):
        return 'AUTOMOBILE'
    else:
        return 'CYBERSECURITY' if 'cyber' in title else 'TECH'  # Categoría por defecto

# Aplicar la función a la columna "title"
df['category'] = df['title'].apply(classify_title)

# Mostrar resumen
print("Resumen de categorías:")
print(df['category'].value_counts())

# Contar cuántas filas quedaron sin categoría válida
valid_categories = ['AI', 'TECH', 'GAMING', 'POLITICS', 'LIFESTYLE', 'SOCIETY', 'AUTOMOBILE', 'CYBERSECURITY']
unclassified_count = df[~df['category'].isin(valid_categories)].shape[0]
print(f"Filas sin categoría válida: {unclassified_count}")

# Guardar el archivo con la nueva columna
output_path = 'archivo_categorizado.csv'
df.to_csv(output_path, sep=',', index=False)
print(f"Archivo guardado en: {output_path}")
```

### Por motivos de organización en la plataforma, vamos a crear un bucle repetitivo usando **colas** para llevar las filas que no contengan datos en la columna "image" al final, dejando al inicio solamente las noticias con imagenes. Esto servirá para tener una mejor estética.

```python
import pandas as pd
from collections import deque

# Cargar el archivo CSV
df = pd.read_csv('archivo_categorizado.csv')

# Crear una cola para las filas sin valores en 'image'
cola_sin_imagen = deque()

# Lista para filas con valores en 'image'
filas_con_imagen = []

# Recorrer cada fila del DataFrame
for _, row in df.iterrows():
    if pd.notna(row['image']) and row['image'] != '':
        filas_con_imagen.append(row)
    else:
        cola_sin_imagen.append(row)

# Agregar las filas de la cola al final de la lista de filas con imagen
while cola_sin_imagen:
    filas_con_imagen.append(cola_sin_imagen.popleft())

# Crear un nuevo DataFrame con el orden actualizado
df_ordenado = pd.DataFrame(filas_con_imagen)

# Guardar el DataFrame ordenado en un nuevo archivo CSV
df_ordenado.to_csv('estructura_definitiva.csv', index=False)

print("Archivo ordenado y guardado como 'estructura_definitiva.csv'")
```
